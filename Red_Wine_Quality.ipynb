```python
matrix_corr = df1.corr()
matrix_corr
```

```python
plt.figure(figsize=(14, 5))
sns.heatmap(matrix_corr, annot=True)
```

```python
correlation = df1['fixed acidity'].corr(df1['quality'])
print(f"Correlation between fixed acidity and quality: {correlation}")

correlation = df1['volatile acidity'].corr(df1['quality'])
print(f"Correlation between volatile acidity and quality: {correlation}")
```

# **These correlations suggest that volatile acidity might be a more important factor in determining quality than fixed acidity.**

```python
plt.figure(figsize=(14, 7))
plt.scatter(df1['fixed acidity'], df1['quality'], c=df1['volatile acidity'], cmap='viridis')
plt.title('Fixed Acidity vs Quality')
plt.xlabel('Fixed Acidity')
plt.ylabel('Quality')
plt.colorbar(label='Volatile Acidity')
plt.show()
```

```python
# Correlation analysis
correlation = df1['volatile acidity'].corr(df1['quality'])
print(f"Correlation between volatile acidity and quality: {correlation}")

plt.figure(figsize=(10, 6))
plt.scatter(df1['volatile acidity'], df1['quality'], alpha=0.7)
plt.title('Volatile Acidity vs. Wine Quality')
plt.xlabel('Volatile Acidity')
plt.ylabel('Quality')
plt.grid(True)
plt.show()
```

# **Higher volatile acidity tends to be associated with lower quality.**

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='quality', y='volatile acidity', data=df1)
plt.title('Wine Quality vs. Volatile Acidity')
plt.xlabel('Wine Quality')
plt.ylabel('Volatile Acidity')
plt.show()
```
